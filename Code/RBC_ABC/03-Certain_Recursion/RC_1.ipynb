## 1. 简单RC模型

### 稳态

```julia
θ = 0.36
δ = 0.1
β = 0.98
((1/β - 1 + δ)/θ)^(1/(θ-1))
```

### 消费函数

```julia
# 消费
function get_c(k1, k2)
    θ = 0.36
    δ = 0.1
    return k1^θ + (1-δ)*k1 - k2
end
```

### 初始化参数

```julia
θ = 0.36
δ = 0.1
β = 0.98
k1 = collect(0.01:0.01:8)
k2 = collect(0.01:0.01:8)
PF = zero(k1)
V1 = zero(k1)
V2 = zero(k1)
wi = zero(k1)
n = length(k1)
```

### 一次迭代

```julia
for i in 1:n
    for j in 1:n
        c = get_c(k1[i], k2[j])
        if c <= 0
            wi[j] = -1e10
        else
            wi[j] = log(c) + β * V2[j]
        end
    end
    V1[i] = maximum(wi)
    PF[i] = k2[argmax(wi)]
end
```

### 多次迭代

```julia
θ = 0.36
δ = 0.1
β = 0.98
k1 = collect(0.01:0.01:8)
k2 = collect(0.01:0.01:8)
PF = zero(k1)
V1 = zero(k1)
V2 = zero(k1)
wi = zero(k1)
n = length(k1)

m = 240
for mm in 1:240
    V2 = copy(V1)
    for i in 1:n
        for j in 1:n
            c = get_c(k1[i], k2[j])
            if c <= 0
                wi[j] = -1e10
            else
                wi[j] = log(c) + β * V2[j]
            end
        end
        V1[i] = maximum(wi)
        PF[i] = k2[argmax(wi)]
    end    
end
```

```julia
using Plots
```

```julia
plot(k1, [k2, PF])
```

## 2. 可变劳动的RC模型

```julia
# 消费
function get_c(k1, k2, h)
    θ = 0.36
    δ = 0.1
    return k1^θ * h^(1-θ) + (1-δ)*k1 - k2
end
```

```julia
θ = 0.36
δ = 0.1
β = 0.98
B = 0.5
k1 = collect(0.01:0.01:10)
k2 = collect(0.01:0.01:10)
h = collect(0:0.01:1)
PFK = zero(k1)
PFH = zero(k1)
V1 = zero(k1)
V2 = zero(k1)
nk = length(k1)
nh = length(h)
wi = zeros(nk, nh);
```

#### 一次迭代

```julia
for i in 1:nk
    for j in 1:nk
        for p in 1:nh
            c = get_c(k1[i], k2[j], h[p])
            if c <= 0
                wi[j, p] = -1e10
            else
                wi[j, p] = log(c) + B*log(1-h[p]) + β * V2[j]
            end
        end
    end
    V1[i] = maximum(wi)
    idx = argmax(wi)
    PFK[i] = k2[idx[1]]
    PFH[i] = h[idx[2]]
end
```

### 多次迭代

```julia
θ = 0.36
δ = 0.1
β = 0.98
B = 0.5
k1 = collect(0.01:0.03:10)
k2 = collect(0.01:0.03:10)
h = collect(0:0.01:1)
PFK = zero(k1)
PFH = zero(k1)
V1 = zero(k1)
V2 = zero(k1)
nk = length(k1)
nh = length(h)
wi = zeros(nk, nh)

m = 20
for mm in 1:m
    V2 = copy(V1)
    for i in 1:nk
        for j in 1:nk
            for p in 1:nh
                c = get_c(k1[i], k2[j], h[p])
                if c <= 0
                    wi[j, p] = -1e10
                else
                    wi[j, p] = log(c) + B*log(1-h[p]) + β * V2[j]
                end
            end
        end
        V1[i] = maximum(wi)
        idx = argmax(wi)
        PFK[i] = k2[idx[1]]
        PFH[i] = h[idx[2]]
    end
end
```

```julia
using Plots
plot(k1, [k2, PFK, PFH])
```
